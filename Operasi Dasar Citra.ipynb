```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import cv2
import numpy as np
```

```python
loadImage = imread(fname="IMAGE.jpg")
loadImagecropped = loadImage.copy()
loadImagecropped = loadImagecropped[0:256,64:320]

fig, axes = plt.subplots(1, 2, figsize = (10, 10))
ax = axes.ravel()

ax[0].imshow(loadImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(loadImagecropped)
ax[1].set_title("Citra Output 2")

fig.tight_layout
plt.show()



```

```python
inv = invert(loadImagecropped)
print('Shape Input : ', loadImagecropped.shape)
print('Shape Output : ',inv.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(loadImagecropped)
ax[0].set_title("CITRA INPUT")

ax[1].hist(loadImagecropped.ravel(), bins=256)
ax[1].set_title('HISTOGRAM INPUT')

ax[2].imshow(inv)
ax[2].set_title('CITRA OUTPUT (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('HISTOGRAM OUTPUT')
```

```python
#salinan dari citra dan mengubah menjadi float 
loadImage = loadImagecropped.copy().astype(float)
#menentukan kecerahan
brightness = 100
loadImage_bright = np.clip(loadImage + brightness, 0, 255).astype(np.uint8)

# figure subplot  
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()
# Plot Gambar Input dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(loadImagecropped, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input")
# Plot Histogram Input
ax[1].hist(loadImagecropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')
# Plot Gambar Output output
ax[2].imshow(cv2.cvtColor(loadImage_bright, cv2.COLOR_BGR2RGB))
ax[2].set_title(f"Citra Output (Brightness={brightness})")
# Plot Histogram output
ax[3].hist(loadImage_bright.ravel(), bins=256, range=(0, 255))
ax[3].set_title('Histogram Output')
```
